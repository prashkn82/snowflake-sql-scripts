```
%%sql -r dataframe_2
select
  seq4() as customer_id,
  'customer_' || (customer_id + 1) as name,
  'customer_' || (customer_id + 1) || '@example.com' as email,
  dateadd(day, -uniform(0, 3650, random()), current_date()) as signup_date,
  round(uniform(100, 100000, random()) / 100, 2) as lifetime_value
from table(generator(rowcount => 1000));
```

```
%%sql -r dataframe_3
SELECT 
    SIGNUP_DATE,
    LIFETIME_VALUE
FROM CUSTOMER;
```

```
# Run the SQL cell above first to populate dataframe_1.
import matplotlib.pyplot as plt
import pandas as pd

dataframe_1['SIGNUP_DATE'] = pd.to_datetime(dataframe_1['SIGNUP_DATE'])
dataframe_1['days_since_signup'] = (pd.Timestamp.today() - dataframe_1['SIGNUP_DATE']).dt.days
dataframe_1 = dataframe_1.sort_values('days_since_signup')

plt.figure(figsize=(10, 6))
plt.scatter(dataframe_1['days_since_signup'], dataframe_1['LIFETIME_VALUE'], alpha=0.3)
plt.plot(dataframe_1['days_since_signup'], dataframe_1['LIFETIME_VALUE'].rolling(100).mean(), color='red', linewidth=2, label='Rolling Avg')

plt.xlabel("Days Since Signup")
plt.ylabel("Lifetime Value")
plt.title("Customer Tenure vs Spend")
plt.legend()
plt.show()
```
